```python
import pandas as pd
import plotly.express as px

df_principal = pd.read_excel('/content/Dados_Imersão_Python.xlsx', sheet_name='PRINCIPAL')
df_principal.head()
```

```python
df_total_acoes = pd.read_excel("/content/Dados_Imersão_Python.xlsx", sheet_name="Total_acoes")
df_total_acoes.head()
```

```python
df_tickers = pd.read_excel("/content/Dados_Imersão_Python.xlsx", sheet_name="TICKERS")
df_tickers.head()
```

```python
df_chat_gpt = pd.read_excel("/content/Dados_Imersão_Python.xlsx", sheet_name="CHAT_GPT")
df_chat_gpt.head()
```

```python
df_principal = df_principal[['Ativo', 'Data', 'ULTVALOR', 'VAR_DIA_PCT']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'ULTVALOR':'valor_final', 'VAR_DIA_PCT':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_tickers, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chat_gpt, left_on='Nome', right_on='Empresas', how='left')
df_principal = df_principal.drop(columns=['Empresas'])
df_principal
```

```python
df_principal['Cat_idade'] = pd.Series(['Mais de 100 anos' if x > 100 else 'Menos de 50 anos' if x < 50 else
                                       'Entre 50 e 100 anos' for x in df_principal['Idade']])
df_principal
```

**Análises**

```python
# Calcule os valores desejados
maior_valor = df_principal['Variacao_rs'].max()
menor_valor = df_principal['Variacao_rs'].min()
media = df_principal['Variacao_rs'].mean()
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

# Imprimir os resultados
print(f"Maior: {maior_valor:,.2f}")
print(f"Menor: {menor_valor:,.2f}")
print(f"Média: {media:,.2f}")
print(f"Média de quem subiu: {media_subiu:,.2f}")
print(f"Média de quem desceu:{media_desceu:,.2f}")
```

```python
df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()
```

```python
# Agrupe os dados pelo segmento e calcule as estatísticas dentro de cada grupo
estatisticas_por_segmento = df_principal.groupby('Segmento').agg({
    'Variacao_rs': ['sum', lambda x: x[x > 0].mean(), lambda x: x[x < 0].mean()]
})

# Renomeie as colunas para corresponder à tabela desejada
estatisticas_por_segmento.columns = ['Variacao_rs', 'Variacao de quem subiu', 'Variacao de quem desceu']

# Reinicie o índice para obter a coluna 'Segmento' como uma coluna regular
estatisticas_por_segmento = estatisticas_por_segmento.reset_index()

estatisticas_por_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
df_analise_saldo
```

**DESAFIO**

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title='Variação Reais por Resultado')
fig.update_layout(yaxis_tickprefix="R$")
fig.update_traces(texttemplate="R$ %{text:,.2f}", textposition="inside")
fig.show()
```

**DESAFIOS**

Análise de Segmento + Gráfico de pizza no df_analise_segmento com a biblioteca Plotly

```python
df_analise_segmento = df_principal.groupby('Segmento')['Variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
#grafico_pizza
fig = px.pie(df_analise_segmento, names='Segmento', values='Variacao_rs', title='Variação em reais por Segmento')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

GroupBy da categoria de idades + gráfico de barras

```python
df_analise_cat = df_principal.groupby('Cat_idade')['Variacao_rs'].sum().reset_index()
df_analise_cat['Variacao_rs'] = df_analise_cat['Variacao_rs'].map('R${:,.2f}'.format)
df_analise_cat
```

```python
fig = px.bar(df_analise_cat, x='Cat_idade', y='Variacao_rs', text='Variacao_rs', title='Qtd. de Empresas por Faixa Etária')
fig.show()
```
